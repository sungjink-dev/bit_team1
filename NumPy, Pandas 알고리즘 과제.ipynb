```
# zeros
def zeros_jin (x,y):
  zeros_arr = []
  for i in range(x):
    row_arr = []
    for j in range(y):
      row_arr.append(0)
    zeros_arr.append(row_arr)
    # print("out for : ",row_arr)
    # print("return : ",zeros_arr)
  return zeros_arr
```

```
# zeros  test data
zeros_jin(3,3)
```

```
# ones
def ones_jin(x,y):
  ones_arr=[]
  for i in range(x):
    row_arr=[]
    for j in range(y):
      row_arr.append(1)
    ones_arr.append(row_arr)
  return ones_arr
```

```
# ones test data
ones_jin(5,5)
```

```
# multiply
def multiply_jin(arr,n):
  multiply_arr=[]
  i = 0; j = 0
  for i in range(len(arr)):
    row_arr=[]
    # if len(arr[i]) > 0: #-> if 절에서 0보다 크면 넣으려고 하는데 에러 걸림..
    for j in range(len(arr[i])):
      row_arr.append(arr[i][j] * n)
    multiply_arr.append(row_arr)
  return multiply_arr
```

```
# multiply test data
arr = [[1,2,3],[4,5,6],[7,8,9]]
n = 4
multiply_jin(arr, n)
```

```
# isnull
def jin_isnull(objs):
  keys = objs.keys()
  values = objs.values()
  objs_TF={}
  for key in keys:
    if objs[key] == None:
      objs_TF.update({key:False})
      # print("T : ",objs_TF)
    else:
      objs_TF.update({key:True})
      # print("F : ",objs_TF)
  return print("check_isnull : ",objs_TF)
```

```
# isnull test data
objs = {'Ohio' : 35000, 'Texas' : None, 'Oregon' : 16000,
          'Utah' : None, 'Michigan' : 4000}
print("objs dict : ", objs)

jin_isnull(objs)
```

```
# transpose
def jin_transpose(data):

# 기존 데이터
  keys = list(data.keys()) # ['Nevada', 'Ohio']
  values = list(data.values()) # [2001, 2002]
  key_keys = []
  
# 변경된 데이터
  data_T = {}

  for i in keys:
    temp_k = list(data[i].keys())
    key_keys += temp_k

  key_keys = list(set(key_keys)) # [2000, 2001, 2002]
#--------------------------------------------------------------------
  for i in key_keys:
    temp_data_T = {i:{}}
    data_T.update(temp_data_T) # data_T = {2000: {}, 2001: {}, 2002: {}}

    for j in keys:
      # print(i,j) # ex. 2000 Nevada
      data_T_inner_dict = {j : 0} # 2000 : 의 value로 넣을 예정
      data_T_key = list(data_T.keys()) # 변경된 데이터의 첫 단계 키 리스트 만들기

      for p in range(len(data_T_key)):
        for q in range(len(key_keys)):
          if data_T_key[p] == key_keys[q]: # 기존 데이터의 키와 변경된 키의 값이 같다
            data_T_inner_dict[j] = data[j][key_keys[q]]
            # print('data_T_inner_dict',data_T_inner_dict)
      temp_data_T[i].update(data_T_inner_dict) # {2000: {'Nevada': 0, 'Ohio': 0}, 2001: {'Nevada': 0, 'Ohio': 0}, 2002: {'Nevada': 0, 'Ohio': 0}}
    print(temp_data_T)
  print(data_T)
```

```
# transpose test data
data = {'Nevada': {2001:2.4, 2002:2.7, 2000:0.},
       'Ohio'  : {2000:1.5,2001:1.7,2002:3.6}}
data
jin_transpose(data)
```
